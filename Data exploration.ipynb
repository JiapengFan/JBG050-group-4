```python
#eval(data[[product]].loc[row index][get list])[tuple in list][value in tuple]
def last_days_dict(data):
    """"
    Creates dictionary of the total remaining amount of products that had to be sold before next
    delivery per product.
    """
    dict = {}
    for a in range(0, len(data.columns[2:len(inventory_df)])):
        product = data.columns[a+2]
        last_chance = 0
        for i in range(0, (len(data)-2)):
            for x in range(0, len(eval(data[[product]].loc[i][0]))):
                if eval(data[[product]].loc[i][0])[x][0] <= (data['day'][i+2]):
                    last_chance =+ eval(data[[product]].loc[i][0])[x][1]
        dict[product] = last_chance
    return dict
```

```python
#eval(data[[product]].loc[row index][get list])[tuple in list][value in tuple]
def last_day_dict(data):
    """"
    Creates dictionary of the total remaining amount of products that had
    to be sold on day of inventory chekc.
    """
    dict = {}
    for a in range(0, len(data.columns[2:len(inventory_df)])):
        product = data.columns[a+2]
        last_chance = 0
        for i in range(0, (len(data)-2)):
            for x in range(0, len(eval(data[[product]].loc[i][0]))):
                if eval(data[[product]].loc[i][0])[x][0] == (data['day'][i]):
                    last_chance =+ eval(data[[product]].loc[i][0])[x][1]
        dict[product] = last_chance
    return dict
```
