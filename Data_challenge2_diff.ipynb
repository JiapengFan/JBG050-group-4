```python
def total_products(data): #takes the dataframe of products before the delivery
    df = data
    
    for s in df.loc[:,'before or after delivery':]:
        for p in df.index:
            string = df[s][p]
            list_string = re.findall(r'\d+', str(string))
            list_int = list(np.int_(list_string))
            relevant_values = [i for i in list_int if list_int.index(i)/2 - list_int.index(i)//2 != 0]
            total = sum(relevant_values)
            df[s][p] = total
            
    return df
```

```python
def total_products_new(data):
    new_data = {}
    days = data['day']
    new_data['day'] = days
    
    for product in data.loc[:,'before or after delivery':]:
        total_list = []
        
        for idx in data.index:
            string = data[product][idx]
            list_string = re.findall(r'\d+', str(string))
           
            list_int = list(np.int_(list_string))

            relevant_indexes = [i for i in range(len(list_int)) if i % 2 > 0]
            relevant_values = [list_int[i] for i in relevant_indexes]
            
            total = sum(relevant_values)
            total_list.append(total)
            
        new_data[product] = total_list
    
    new_df = pd.DataFrame(new_data)
    return new_df
    
```

```python
def product_difference(before, after):
    after1 = after.set_index('day')
    before1 = before.set_index('day')
    
    df = after1.copy()
    
    for product in after1.columns:#.loc[:,'Komkommer':]:
        for day in range(len(after1)-1):
            day_after = after1.index[day]
            day_before = after1.index[day + 1]
            difference = ((after1[product][day_after]) - (before1[product][day_before]))
            df[product][day_after] = difference
            
            
    return df
```

```python
def new_product_difference(before, after):
    new_before = before[1:]
    new_after = after[:-1]
    
    idx = new_after['day']
    new_df = pd.DataFrame()
    new_df['day'] = idx
    new_df.set_index('day')
    
    for product in new_after.columns[2:]:
        diff_list = []
        for day in range(len(new_after)):
            diff = (new_after[product].iloc[day] - new_before[product].iloc[day])
            diff_list.append(diff)
            
        new_df[product] = diff_list
        
    return new_df
```

```python
difference = new_product_difference(total_products_before, total_products_after)
difference
```
